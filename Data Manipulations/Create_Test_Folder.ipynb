```

import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from joblib import Parallel, delayed
import multiprocessing
from random import shuffle
import shutil
```

```
TEST_PRECENTAGE = 0.2

DATADIR = "drive/My Drive/photos/"
TESTDIR = "drive/My Drive/test/"

CATEGORIES = os.listdir(DATADIR);
   
def create_test_folder(category):
    path = os.path.join(DATADIR, category)
    new_path = os.path.join(TESTDIR, category)
    
    if not os.path.exists(new_path):
        os.makedirs(new_path)  
        
    list = os.listdir(path)
    shuffle(list)
    
    test_size = TEST_PRECENTAGE * len(list);    
    test_landmarks = list[:int(test_size)];
    
    print("Landmark Id : ", category , ". Moving ", len(test_landmarks), " Photos")
    
    for photo_name in test_landmarks:
      old_photo_loc = path + "/" + photo_name;
      new_photo_loc = new_path + "/" + photo_name
      os.rename(old_photo_loc, new_photo_loc);
      

    
    print("Done category " + category);
    
for category in CATEGORIES:
  create_test_folder(category);
```

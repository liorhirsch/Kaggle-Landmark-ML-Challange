```

import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from joblib import Parallel, delayed
import multiprocessing
from random import shuffle
```

```
TEST_PRECENTAGE = 0.05

DATADIR = "drive/My Drive/photos/"
TESTDIR = "drive/My Drive/smallDS/"

CATEGORIES = os.listdir(DATADIR);
   
def create_test_folder(category):
    path = os.path.join(DATADIR, category)
    new_path = os.path.join(TESTDIR, category)
    
    if not os.path.exists(new_path):
        os.makedirs(new_path)  
        
    list = os.listdir(path)
    shuffle(list)
    
    test_size = TEST_PRECENTAGE * len(list);    
    test_landmarks = list[:int(test_size)];
    
    print("Landmark Id : ", category , ". Moving ", len(test_landmarks), " Photos")
    
    for photo_name in test_landmarks:
      old_photo_loc = path + "/" + photo_name;
      new_photo_loc = new_path + "/" + photo_name
      os.rename(old_photo_loc, new_photo_loc);
    
    print("Done category " + category);
    
for category in CATEGORIES:
  create_test_folder(category);
print("Done!")
```

# Move back the pictures to the dataset's folder

```
from shutil import copyfile
```

```
TO_DIR = "drive/My Drive/photos/"
FROM_DIR = "drive/My Drive/smallDS/"

CATEGORIES = os.listdir(FROM_DIR);

total_handled_photos = 0;

def spread_back_photos(category):
    old_path = os.path.join(FROM_DIR, category)
    new_path = os.path.join(TO_DIR, category)
    
    if not os.path.exists(new_path):
        print("path %s not found, creating it..", new_path)
        os.makedirs(new_path)  
            
    filesToCopy = os.listdir(old_path)
    
    print("Landmark Id : ", category , ". Coping ", len(filesToCopy), " photos from ", old_path, " to ", new_path)
    
    for photo_name in filesToCopy:
      old_photo_loc = old_path + "/" + photo_name
      new_photo_loc = new_path + "/" + photo_name
      copyfile(old_photo_loc, new_photo_loc)
#       os.rename(old_photo_loc, new_photo_loc); # IF YOU WANT TO CUT THE PHOTOS INSTEAD OF JUST COPY THEM BACK - COMMENT THE LINE BEFORE AND vUNCOMMENT THIS
      total_handled_photos += 1

    
index = 1;    
for category in CATEGORIES:
  spread_back_photos(category);
  print(index, "/", len(CATEGORIES), " Done category " + category);
  index += 1;
print("Done! total ", total_handled_photos, " photos.")
```
